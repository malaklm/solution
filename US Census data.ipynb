### This code works well but there's a new problem. The Fact column have many values separated by (,)  you can deal with it later so we just ignore the column and take all columns between Alabama and Wyoming. <br>

during I solve this problem I saw there's the value of D and : in the columns so I just add it in the filter

```python
df_census.loc[:,'Alabama':'Wyoming'].replace({",|\%|\$|\"|Z|\s|\D|\:": ''}, regex=True,inplace=True)
df_census.head()
```

```python
# now you can convert the columns type
df_census.loc[:,'Alabama':'Wyoming'] = df_census[["Arizona", "Wyoming"]].apply(pd.to_numeric)
df_census.dtypes
```
